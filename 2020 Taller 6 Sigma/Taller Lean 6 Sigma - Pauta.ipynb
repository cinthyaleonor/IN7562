# Lean Six Sigma : el inicio $\dots$
<font size="5">
<ul>
    <li> El concepto de Lean Six Sigma se publicó por primera vez en 2002 en el libro de Michael George y Peter Vincent Lean Six Sigma: Combinando Six Sigma con Lean Speed.
    <li> Se basa en que las actividades que generan criticidad de calidad para el cliente y las que producen los mayores retrasos en un proceso, ofrecen oportunidades para mejorar en costo, calidad, capital y tiempos de entrega, 
    <li> La metodología incluye fases para eliminar el desperdicio (Lean) y las herramients de Six Sigma para gestión de calidad 
</ul>

</font>

#  Lean Six Sigma : Definición General

|**Lean**  | **Six Sigma**  |
| :--------| :--------------|
|Es una metodología que se enfoca en eliminar actividades innecesarias que no generan valor durante la creación de un producto o servicio, quedándose con el flujo de procesos óptimo, completo, racionalizado y eficaz que logre satisfacer a cada cliente. |Es una metodología de mejora continua que busca calidad a través de la reducción de defectos en cada proceso para reducir costos e incrementar la satisfacción de los clientes. Six Sigma se enfoca en reducir la variación de los procesos para no generar más de 3.4 defectos por millón. |

**tl;dr : Reducir Desperdicios + Evitar Errores = LEAN 6 Sigma**

```python
# Importamos las librerías esenciales
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
import seaborn as sns
sns.set_palette(sns.diverging_palette(10, 220, sep=80, n=7))

#Obtenemos los datos
df =  pd.read_csv("https://raw.githubusercontent.com/cinthyaleonor/IN7562/master/2020%20Taller%206%20Sigma/data_glotontena.csv", sep =";")
df['fecha'] = pd.to_datetime(df['fecha'], format='%d-%m-%Y')
```

```python
# Miremos nuestros datos numéricos 
df.describe().transpose().round(1)
#Agrupamos datos por fecha
```

```python
# Creamos una figura y el espacio plot
plt.subplots(figsize=(15,5))
plt.plot(df.groupby('fecha')['t_despacho'].mean(), label="T° Promedio Despacho", color = 'crimson')
plt.plot(df.groupby('fecha')['t_preparacion'].mean(), label="T° Promedio Preparación", color = 'cadetblue')
plt.title("T° promedio de despacho", fontsize=14, fontweight='bold')
plt.suptitle("La Glotontena", fontsize=10)
plt.legend(loc='center right') ; plt.xlabel("Fecha") ; plt.ylabel("Tiempo despacho (minutos)") ; plt.show()
```

```python
#Baseline CTQ para marzo
promedio_m = df.loc[df['fecha'].dt.month == 4, "t_despacho"].mean()
df_marzo = df.loc[df['fecha'].dt.month == 4, ["fecha","t_despacho"]]
#df_marzo.fecha.dt.day
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,2), sharey=True, dpi=120)
ax1.plot(df_marzo.t_despacho, label="T° Despacho Marzo", color = 'crimson')
ax1.axhline(promedio_m, color='b', linestyle='dashed') 
ax2.plot(df.groupby(df_marzo.fecha.dt.day)['t_despacho'].mean(), label="T° Promedio Despacho", color = 'crimson')
# Nombre ejes 
ax1.set_title('Tiempo despacho marzo'); ax2.set_title('Tiempo promedio diario despacho marzo')
ax1.set_xlabel('N° Pedido');  ax2.set_xlabel('Dia del Mes')  # x label
ax1.set_ylabel('T° (minutos)');  ax2.set_ylabel('T° (minutos)')  # y label
```

```python
##### Distribución Tiempos de orden por producto 
f, axes = plt.subplots(ncols=2, figsize=(15, 5), sharex=False)
sns.distplot( df["t_despacho"] , color="crimson", ax=axes[0]) #; axes[0].set_xlim((10,100)) 
sns.distplot( df["t_preparacion"] , color="cadetblue", ax=axes[1]) #; axes[1].set_xlim((10,25)) 
plt.show()
```

```python
# Revisemos la situación relacionada con las transacciones fallidas
labels = {1:"Correcta",0:"Fallida"}
per = pd.DataFrame({ "transaccion": df.estado_transaccion.unique(),
                    "num" : df["estado_transaccion"].value_counts(),
                     "porcentaje" : df["estado_transaccion"].value_counts(normalize=True)*100 })
per.transaccion = per.transaccion.map(labels)
per
```

```python
#Scatter estilo 1 
sns.pairplot(df[['menu','t_despacho','monto','calificacion']] ,
             hue="menu", kind = "scatter", 
             plot_kws= dict(alpha=0.8)) #,palette = "husl"
plt.show()
```

```python
#Analizamos el comportamiento por menu
temp = pd.merge(df.groupby("menu")["monto"].sum() , 
                df.groupby("menu")["monto"].count(),
                on = "menu")
temp
menu = pd.DataFrame({"menu" : temp.index.values, 
                     'monto': temp.monto_x,
                     'n_pedidos': temp.monto_y} )
                    
menu["calificacion_m"] = round(df.groupby("menu")["calificacion"].mean(),2)
menu["monto_medio_pp"] = round(menu.monto/menu.n_pedidos,2)
menu.sort_values( by = ['n_pedidos'] , inplace=True , ascending =False )
# menu.reset_index(drop = True, inplace=True )
menu
```

```python
#Análisis de la regla de Pareto >> Creamos función para graficar
def pareto_plot(df, x=None, y=None, title=None, show_pct_y=False, pct_format='{0:.0%}'):
    xlabel = x
    ylabel = y
    tmp = df.sort_values(y, ascending=False)
    x = tmp[x].values
    y = tmp[y].values
    weights = y / y.sum()
    cumsum = weights.cumsum()
    
    fig, ax1 = plt.subplots()
    ax1.bar(x, y)
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(ylabel)

    ax2 = ax1.twinx()
    ax2.plot(x, cumsum, '-ro', alpha=0.5)
    ax2.set_ylabel('', color='r')
    ax2.tick_params('y', colors='r')
    
    vals = ax2.get_yticks()
    ax2.set_yticklabels(['{:,.2%}'.format(x) for x in vals])

    # hide y-labels on right side
    if not show_pct_y:
        ax2.set_yticks([])
    
    formatted_weights = [pct_format.format(x) for x in cumsum]
    for i, txt in enumerate(formatted_weights):
        ax2.annotate(txt, (x[i], cumsum[i]), fontweight='heavy')    
    
    if title:
        plt.title(title)
    
    plt.tight_layout()
    plt.show()
```

```python
#Analizamos el porcetanje de pedidos acumulados 
pareto_plot( menu.loc[:,['menu','n_pedidos']] , x='menu', y='n_pedidos', title=None, show_pct_y=False, pct_format='{0:.0%}')
```

```python
# Analizamos el % acumulado del monto total por tipo tipo de menú
pareto_plot( menu.loc[:,['menu','monto']] , x='menu', y='monto', title=None, show_pct_y=False, pct_format='{0:.0%}')
```
